```python
# do above for all input content portrait images
directory = './imgs/input/'
for basename in os.listdir(directory):
    if basename.endswith(".jpg") or basename.endswith(".png"): 
        filename = directory + basename
        print(filename)
        img = io.imread(filename)
        mask = get_features_mask(img)
        new_dir = './imgs/face_masks_gaussian/'
        new_filename = new_dir + basename[:-4] + '_gaussian_face_mask.jpg'
        print(new_filename)
        imageio.imwrite(new_filename, mask)
        
        
```

```python
# don't enlarge the facial polygons
directory = './imgs/input/'
for basename in os.listdir(directory):
    if basename.endswith(".jpg") or basename.endswith(".png"): 
        filename = directory + basename
        print(filename)
        img = io.imread(filename)
        mask = get_features_mask(img, doEnlarge=False)
        new_dir = './imgs/face_masks_gaussian_noEnlarge/'
        new_filename = new_dir + basename[:-4] + '_gaussian_face_mask_noEnlarge.jpg'
        print(new_filename)
        imageio.imwrite(new_filename, mask)
        
        
```
