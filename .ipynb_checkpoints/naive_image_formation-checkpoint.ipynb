```python
import face_alignment
from skimage import io
import matplotlib.pyplot as plt;
import numpy as np
```

```python
# Load pastiche
pastiche = './gout_20.jpg'
pastiche_data = io.imread(pastiche)
plt.imshow(pastiche_data)
plt.show()
```

```python
# Load masks
fgbg_mask = './imgs/masks/lupita_mask.png'
fgbg_mask_data = io.imread(fgbg_mask)[:, :, :3]
fgbg_mask_data.astype(bool)
plt.imshow(fgbg_mask_data)
plt.show()

face_mask = './imgs/face_masks_gaussian/lupita_gaussian_face_mask.jpg'
face_mask_data = io.imread(face_mask)
plt.imshow(face_mask_data)
plt.show()
```

Now we apply the masks to retain parts of the image.

```python
# Background
pastiche_bg = np.where(fgbg_mask_data==False, pastiche_data, 0)
plt.imshow(pastiche_bg)
plt.show()
```

```python
# Foreground - Facial Features
pastiche_fg = np.where(fgbg_mask_data==True, pastiche_data, 0)
plt.imshow(pastiche_fg)
plt.show()
```
