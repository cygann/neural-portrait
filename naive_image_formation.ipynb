```python
import face_alignment
from skimage import io
import matplotlib.pyplot as plt;
import numpy as np
```

```python
# Load gatys image
gatys = './imgs/gatys/s6baseline.jpg'
gatys_data = io.imread(gatys)
plt.imshow(gatys_data)
plt.show()
```

```python
# Load gatysplus image
gatysplus = './imgs/gatysplus/lupitas6gatysplus.jpg'
gatysplus_data = io.imread(gatysplus)
plt.imshow(gatysplus_data)
plt.show()
```

```python
# Load foreground mask
fgbg_mask = './imgs/masks/lupita_mask.png'
fgbg_mask_data = io.imread(fgbg_mask)[:, :, :3]
fgbg_mask_data.astype(int)
plt.imshow(fgbg_mask_data)
plt.show()
```

Now we apply the mask to make the foreground less stylized than the background.

```python
naive = np.where(fgbg_mask_data>0, gatys_data, gatysplus_data)
plt.imshow(naive)
plt.show()
```

```python
io.imsave('./imgs/naive/lupitas6_naive.jpg', naive)
```
